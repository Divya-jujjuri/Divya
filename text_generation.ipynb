```python
from transformers import pipeline
```

```python
a = pipeline("sentiment-analysis")
```

```python
text =input("Enter the text: ")
print(a(text))
```

```python
gen = pipeline("text-generation")
```

```python
text = input("enter the text: ")
t = gen(text, max_length =200)
```

```python
print(t[0].get('generated_text'))
```

```python
text = input("Enter a text")
result = a(text)
result
```

```python
qa = pipeline('question-answering')
```

```python
c = '''bitty some butter,but the butter was bitter,so bitty bought some better butter to make her bitter butter better'''
```

```python
q = "what does bitty bought"
```

```python
result =qa({'question': q,'context':c})
print(result)
```

```python
s = pipeline('summarization')
```

```python
t = ''' Computer vision is a field of artificial intelligence (AI) that focuses on enabling computers to "see" and interpret the visual world, just like humans do. It involves using algorithms and models to extract information from images and videos, allowing computers to identify objects, detect patterns, and make decisions based on what they see. '''
```

```python
result = s(t)
result[0].get('summary_text')
```
